```python
import chromadb

# Data stored at a path.
client = chromadb.PersistentClient(path="./my_database")
collection = client.create_collection(name="Boys")
collection = client.get_collection("Boys")
collection = client.get_or_create_collection(
    name="Boys"
)
collections = client.list_collections()

print(collections)
```

```python
import chromadb

# Create persistent client
client = chromadb.PersistentClient(path="./chroma_db")

# Create collection
collection = client.get_or_create_collection("books")

# Add documents
collection.add(
    ids=["1", "2"],
    documents=[
        "Artificial Intelligence is transforming healthcare.",
        "Deep Learning powers modern AI systems."
    ],
    metadatas=[
        {"category": "AI", "year": 2025},
        {"category": "ML", "year": 2024}
    ]
)

# Query with metadata filter
results = collection.query(
    query_texts=["AI in medicine"],
    where={"category": "ML"},
    n_results=2
)

print(results)
```

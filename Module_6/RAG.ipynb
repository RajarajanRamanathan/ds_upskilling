```python
RAG (Retrieval-Augmented Generation)
What is RAG?

Retrieval-Augmented Generation (RAG) is an AI architecture that combines:

Information Retrieval → Finds relevant documents from an external knowledge base.
Large Language Model (LLM) → Uses those retrieved documents to generate an accurate answer.

Instead of relying only on the knowledge stored in the model parameters, RAG allows the model to access up-to-date, private, or domain-specific information at inference time.

Architecture
                User Question
                      │
                      ▼
              Query Embedding
                      │
                      ▼
            Vector Database Search
                      │
        Top-K Relevant Documents
                      │
                      ▼
        Prompt + Retrieved Context
                      │
                      ▼
                 Large Language Model
                      │
                      ▼
                Final Answer
```

```python
Core problem RAG solves:
1. Hallucination - The model may generate confident but incorrect information.
2. Outdated Knowledge - LLMs only know information from their training data.
3. Private Knowledge - Companies have internal documents:
HR policies
Medical records
Financial reports
Technical documentation

Training an LLM on these documents is often impractical or undesirable.
4. Large Knowledge Bases - The LLM cannot memorize all of them.(10M Pdf)
```

```python
Indexing Pipeline:	Ingest → Parse → Chunk → Embed → Store (offline preparation).
Query Pipeline:	Query → Embed → Retrieve → Rerank → Generate (online inference).
Naive RAG:	Basic vector retrieval followed by generation.
Advanced RAG:	Adds techniques like hybrid search, query rewriting, and reranking for better accuracy.
Modular RAG:	Treats each stage as an interchangeable component for flexible, production-scale systems.
```
